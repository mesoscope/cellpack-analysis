## Get distribution of counts and sizes from manifest

```python
import pandas as pd
import numpy as np
```

```python
meta_df = pd.read_csv("/allen/aics/animated-cell/Saurabh/cellpack-analysis/data/variance_dataset.csv", index_col="CellId")
```

```python
for gene in ["SLC25A17", "RAB5A"]:
    df_gene = meta_df.loc[meta_df["structure_name"] == gene]
    counts = df_gene["STR_connectivity_cc"].values
    volume_per_unit = df_gene["STR_shape_volume"].values / counts
    unit_radius = (volume_per_unit / (4/3 * np.pi))**(1/3)
    print(gene)
    print(f"Count: {np.mean(counts):0.2f} +/- {np.std(counts):.2f}")
    print(f"Volume: {np.mean(volume_per_unit):0.2f} +/- {np.std(volume_per_unit):.2f}")
    print(f"Radius: {np.mean(unit_radius):0.2f} +/- {np.std(unit_radius):.2f}")
```
